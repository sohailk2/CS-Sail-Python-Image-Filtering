```python
#Import Statements
from PIL import Image
import numpy as np
import scipy.misc as smp
import random
```

### Solutions for the Google Blue Color Change

```python
googleImage = Image.open("Images/googleLogo.png")
googleBitmap = np.array(googleImage)

for i in range(len(googleBitmap)):
    for j in range(len(googleBitmap[i])):
        pixel = googleBitmap[i][j]
        
        if (pixel[0] < 255 and pixel[1] < 255 and pixel[2] > 150):
            #after identifying all blue pixels we changed the rgb values to
            # (255,130,198)
            pixel[0] = 255
            pixel[1] = 130
            pixel[2] = 198
        
result = Image.fromarray(googleBitmap)
display(result)
```

### Solutions to change the background

```python
googleImage = Image.open("Images/googleLogo.png")
googleBitmap2 = np.array(googleImage)

for i in range(len(googleBitmap2)):
    for j in range(len(googleBitmap2[i])):
        pixel = googleBitmap2[i][j]
            
        #we found all the white values in the image. white has an rgb of (255,255,255)
        if (pixel[0] == 255 and pixel[1] == 255 and pixel[2] == 255):
            #now we are se
            pixel[0] = 100
            pixel[1] = 100
            pixel[2] = 100
        
result = Image.fromarray(googleBitmap2)
display(result)
```

### Solutions to Stenopgraphy

#### How to encode message

```python
messageImage = Image.open("Images/Secret.png")
display(messageImage)
```

### How we created the image

```python
#load the pixels from the image object into a 2d array
messageBitmap = np.array(messageImage)

for i in range(len(messageBitmap)):
    #iterate through every pixel in the array
    for j in range(len(messageBitmap[i])):
        #randomly generate a boolean
        flag = random.choice([True, False])
        
        pixel = messageBitmap[i][j]
        
        #if the pixel is not the color of the message
        if not (pixel[0] == 127 and pixel[1] == 127 and pixel[2] == 127):
            #if we randomly generate true, set the pixel to black
            if (flag):
                pixel[0] = 0
                pixel[1] = 0
                pixel[2] = 0
            #if we generate false, set the pixel to white
            else:
                pixel[0] = 255
                pixel[1] = 255
                pixel[2] = 255

#save the result as an image object
result = Image.fromarray(messageBitmap)
#save the image object to a file
result.save("Images/EncodedSecret.png")
```

### Solution to decode image

```python
secretImage = Image.open("Images/EncodedSecret.png")
```

```python
#load the pixels from the image object into a 2d array
secretBitmap = np.array(secretImage)


for i in range(len(secretBitmap)):
    #iterate through every pixel in the array
    for j in range(len(secretBitmap[i])):        
        pixel = secretBitmap[i][j]
        
        #if the pixel is black, set it to white to reveal the image
        if (pixel[0] == 0 and pixel[1] == 0 and pixel[2] == 0):
            pixel[0] = 255
            pixel[1] = 255
            pixel[2] = 255
    
#save the result as an image object
result = Image.fromarray(secretBitmap)
#display the image object
display(result)
```
